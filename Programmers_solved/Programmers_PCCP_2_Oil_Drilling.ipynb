### 정답 코드

```python
from collections import deque

def solution(land):
    col = len(land)
    row = len(land[0])

    visit = [[0 for r in range(row)] for c in range(col)]

    dx = [1, 0, -1, 0]
    dy = [0, 1, 0, -1]
    oil_info = list()
    amount_list = [0 for x in range(row)]
    for c in range(col):
        for r in range(row):
            if land[c][r] == 0 or visit[c][r]:
                continue
            oil = 0
            visit[c][r] = 1
            queue = deque()
            local_set = set()
            local_set.add(r)
            queue.append((c,r))
            while(len(queue) != 0):
                oil+=1
                y,x = queue.popleft()
                local_set.add(x)
                for cursor in range(4):
                    nx = x + dx[cursor]
                    ny = y + dy[cursor]
                    if nx < 0 or nx >= row or ny < 0 or ny >= col:
                        continue
                    if land[ny][nx] == 0 or visit[ny][nx]:
                        continue
                    visit[ny][nx] = 1
                    queue.append((ny,nx))
            oil_info.append((local_set,oil))

    for i in range(len(oil_info)):
        s,o = oil_info[i]
        for r in list(s):
            amount_list[r] += o
    return max(amount_list)
```
